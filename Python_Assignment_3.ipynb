Assignment

Calculate f(x) repetitive by using input command. Once, 0 is used as an input stop the program

image.png

```python
def f(x):
    if x == 0:
        result = "i cannot calculate with arguments < 0"
    else:
        result = x**2 + 3
    return result

while True:
    x = int(input("Type in a number: "))
    if x != 0:
        print(f(x))
    else:
        print("End of Program")
        break
```
